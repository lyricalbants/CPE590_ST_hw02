#### Question 1

The Cost function, J is given by $arg_wmin[||\bold{y-xw}||^{2} + \lambda||\bold{w}||^2]$

Expanding the first expression:

\begin{align*}
||\bold{y-xw}||=(\bold{y-xw})^T(\bold{y-xw}) \\
\bold{y}^T\bold{y}-\bold{y}^T\bold{x}\bold{w}-\bold{y}\bold{x}^T\bold{w}^T+\bold{x}^T\bold{x}\bold{w}^T\bold{w} \\
\end{align*}
since $y^Tx$ is scalar (from dot product), and $(\bold{y}\bold{x}^T)^T$ is the transpose of $\bold{y}^T\bold{x}$, so $\bold{y}^T\bold{x}=\bold{yx}^T$
Then, 

\begin{align}
J = \bold{y}^T\bold{y}-2\bold{y}^T\bold{x}\bold{w}+\bold{x}^T\bold{x}\bold{w}^T\bold{w} + \lambda\bold{w}^T\bold{w}
\end{align}
Differentiation with respect to w
\begin{align*}
\frac{\delta J}{\delta \bold{w}}=-2\bold{x}^T\bold{y} + 2\bold{x}^T\bold{x}\bold{w} + 2\lambda\bold{w}
\end{align*}
Equating the expression to zero to solve for w and dividing by 2
\begin{align*}
0 = -\bold{yx}^T + \bold{x}^T\bold{x}\bold{w} + \lambda\bold{w} \\
\bold{x}^T\bold{x}\bold{w} + \lambda\bold{w} = \bold{yx}^T \\
\bold{w}(\bold{x}^T\bold{x} + \lambda\bold{I}) = \bold{yx}^T \\
\bold{w} = (\bold{x}^T\bold{x} + \lambda\bold{I})^{-1}\bold{yx}^T
\end{align*}
Where $\bold{I}$ is the identity matrix

Compared to the original linear regression solution, $\bold{w}_{ols}=(\bold{x}^T\bold{x})^{-1}\bold{x}^T\bold{y}$, this regularization adds a new term $(\lambda\bold{I})^{-1}$ to the OLS.
If $\lambda$ is 0, it reduces to the original linear regression solution, if it is large, it shrinks $\bold{w}$ preventing overfitting.

Lagrange Multiplier is $L(x,y,\lambda)=f(x,y)-\lambda g(x,y)\:where\:g(x,y)=x^3+y^3-16=0$

$\nabla f = [\frac{\delta f}{\delta x} \: \frac{\delta f}{\delta y}] = [ye^{xy} \: xe^{xy}]$, $\nabla g = [\frac{\delta g}{\delta x} \: \frac{\delta g}{\delta y}] = [3x^2 \: 3y^2]$

\begin{align*}
\frac{\delta f}{\delta x} - \lambda \frac{\delta g}{\delta x} = 0 \\
ye^{xy} - \lambda 3x^2 = 0 \\
\frac{\delta f}{\delta y} - \lambda \frac{\delta g}{\delta y} = 0 \\
xe^{xy} - \lambda 3y^2 = 0 \\
\end{align*}

With three equations, we can solve for the $x, y, \lambda$
\begin{align}
ye^{xy} = \lambda 3x^2 \\
xe^{xy} = \lambda 3y^2 \\
x^3 + y^3 = 16 \\
\end{align}

\begin{align*}
\frac{ye^{xy}}{xe^{xy}} = \frac{3\lambda x^2}{3\lambda y^2} \\
\frac{y}{x} = \frac{x^2}{y^2} \\
y^3 = x^3
\end{align*}
Hence, $y = x$

\begin{align*}
x^3 + x^3 = 16 \\
2x^3 = 16\\

x = 2
\end{align*}

```python
# Define the objective function
def objective(x):
    x, y = x
    return torch.exp(x*y)

# Define the constraint function
def constraint(x):
    return x[0]**3 + x[1]**3 - 16

# Penalty method: Combine objective and constraint
def loss_fn(x, penalty_weight):
    obj = objective(x)
    constr_violation = constraint(x)**2 
    return obj + penalty_weight * constr_violation


epochs = 5000
penalty_weights = [10, 100, 1000]

# Iterate over different penalty weights
for penalty in penalty_weights:
    # Initialize trainable parameters
    x = torch.tensor([1.0, 1.0], requires_grad=True)

    # Set up the optimizer
    optimizer = torch.optim.SGD([x], lr=0.0001)

    # Perform gradient descent
    for i in range(5000):
        optimizer.zero_grad()  
        loss = loss_fn(x, penalty) 
        loss.backward()  
        optimizer.step() 

        # Print progress (optional)
        if (i + 1) % 1000 == 0:
            print(f"[Penalty={penalty}] Iter {i+1}: x = {x[0].item():.4f}, y = {x[1].item():.4f}, Loss = {loss.item():.4f}")

    # Print the final results for this penalty weight
    print(f"\n[Penalty={penalty}] Optimal values:")
    print(f"x = {x[0].item():.4f}")
    print(f"y = {x[1].item():.4f}")
    print(f"Objective value = {objective(x).item():.4f}")
    print(f"Constraint violation = {constraint(x).item():.4f}\n")
```

```python
# Define the objective function
def objective(x):
    x1, x2 = x
    return 2 * torch.pi * x1 * (x1 + x2)

# Define the constraint function
def constraint(x, V):
    x1, x2 = x
    return torch.pi * x1**2 + x2 - V  # Constraint: 4x^2 + y^2 = 9

# Define the projection function
def project(x, V):
    x1, x2 = x
    # Project onto the constraint surface 4x^2 + y^2 = 9
    scale = torch.sqrt(V / (torch.pi * x1**2 * x2))
    return torch.tensor([x1 * scale, x2 * scale], requires_grad=True)


learning_rate = 0.001
V = [30.0, 40.0, 50.0, 60.0]

# Perform gradient projection
for v in V:

    # Initialize trainable parameters
    x = torch.tensor([1.0, 2.0], requires_grad=True)  # Initial guess

    # Set up the optimizer (gradient descent)
    optimizer = torch.optim.Adam([x], lr=learning_rate)

    for i in range(1000):
        # Zero the gradients
        optimizer.zero_grad()

        # Compute the objective
        loss = objective(x)

        # Backpropagate to compute gradients
        loss.backward()

        # Take a gradient step
        optimizer.step()

        # Project onto the feasible region
        with torch.no_grad():
            x.data = project(x, v)

        # Print progress (optional)
        if (i + 1) % 100 == 0:
            print(f"[V={v}] Iteration {i + 1}: x = {x[0].item():.4f}, y = {x[1].item():.4f}, Loss = {loss.item():.4f}")

    # Print the final results
    print("\nOptimal values:")
    print(f"x = {x[0].item():.4f}")
    print(f"y = {x[1].item():.4f}")
    print(f"Objective value = {objective(x).item():.4f}")
    print(f"Constraint violation = {constraint(x, v).item():.4f}")
```

```python
# Extract features and target
X = toluca["LotSize"].values.reshape(-1, 1)  
y = toluca["WorkHours"].values.reshape(-1, 1)

X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.float32)

# Normalize X
X_mu, X_std = X.mean(), X.std()
X = (X - X_mu) / X_std
y_mu, y_std = y.mean(), y.std()
y = (y - y_mu) / y_std 

w = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)

optimizer = optim.SGD([w, b], lr=0.01)

loss_list = []
w_list, b_list = [], []

for epoch in range(5000):
    optimizer.zero_grad()
    y_pred = w * X + b
    loss = torch.mean((y_pred - y) ** 2)
    loss.backward()
    optimizer.step()
    loss_list.append(loss.item())
    w_list.append(w.item())
    b_list.append(b.item())

    # Print every 100 epochs
    if (epoch + 1) % 100 == 0:
        print(f"Epoch {epoch+1}: Loss = {loss.item():.4f}, w = {w.item():.4f}, b = {b.item():.4f}")

# Q1: Plot Loss vs. Iteration
plt.figure(figsize=(6, 4))
plt.plot(loss_list, label="Loss")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.title("Loss vs. Iteration")
plt.legend()
plt.show()

# Q2: Plot Coefficients vs. Iteration
plt.figure(figsize=(6, 4))
plt.plot(w_list, label="w (Slope)")
plt.plot(b_list, label="b (Intercept)")
plt.xlabel("Iteration")
plt.ylabel("Coefficient Value")
plt.title("Coefficients vs. Iteration")
plt.legend()
plt.show()

# Q3 Scatter Plot + Regression Line
X_plot = X.numpy() * X_std.numpy() + X_mu.numpy()
y_plot = y.numpy() * y_std.numpy() + y_mu.numpy()
y_pred_plot = (w.item() * X + b.item()) * y_std + y_mu

plt.figure(figsize=(6, 4))
plt.scatter(X_plot, y_plot, label="Data")
plt.plot(X_plot, y_pred_plot.detach().numpy(), color="red", label="Fitted Line")
plt.xlabel("Lot Size")
plt.ylabel("Work Hours")
plt.title("Regression Fit")
plt.legend()
plt.show()

# Q4: Predict WorkHours for LotSize = 75 
lot_size = 75
lot_size_norm = (lot_size - X_mu.item()) / X_std.item()
work_hours_pre = (w.item() * lot_size_norm + b.item()) * y_std + y_mu
print(f"Predicted Work Hours for LotSize = 75: {work_hours_pre.item():.4f}")

# Q5: Compute Confidence Interval
y_resid = y - (w * X + b)
sigma = torch.std(y_resid).item()
n = len(X)
X_aug = torch.cat([torch.ones(n, 1), X.view(-1, 1)], dim=1)
H = X_aug @ torch.pinverse(X_aug.T @ X_aug) @ X_aug.T
std_error = sigma * torch.sqrt(torch.diag(H)).view(-1, 1) 


# Compute upper and lower confidence bounds
y_pred_values = (w.item() * X + b.item()) * y_std + y_mu
upper = y_pred_values + (1.96 * std_error)
lower = y_pred_values - (1.96 * std_error)


# Plot confidence bands
plt.figure(figsize=(6, 4))
plt.scatter(X_plot, y_plot, label="Data")
plt.plot(X_plot, y_pred_plot.detach().numpy(), color="red", label="Fitted Line")
plt.fill_between(X_plot.squeeze(),
                 lower.squeeze().detach().numpy(), 
                 upper.squeeze().detach().numpy(),
                 color='gray', alpha=0.3, label="95% Confidence Interval")
plt.xlabel("Lot Size")
plt.ylabel("Work Hours")
plt.title("Regression Fit with Confidence Band")
plt.legend()
plt.show()
```

```python
# Given parameters
k = 1.216e-2 
N0 = 5.0
t_min, t_max = 0.0, 200.0
t_step = 0.004
epochs = 500000
learning_rate = 0.001

t_values = np.arange(t_min, t_max, t_step)
N_analytical = N0 * np.exp(-k * t_values)

t_tensor = torch.tensor(t_values, dtype=torch.float32).view(-1, 1)
N_tensor = torch.tensor(N_analytical, dtype=torch.float32).view(-1, 1)

# Define a polynomial model of degree 3
class PolyModel(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.a = torch.nn.Parameter(torch.randn(1, requires_grad=True))
        self.b = torch.nn.Parameter(torch.randn(1, requires_grad=True))
        self.c = torch.nn.Parameter(torch.randn(1, requires_grad=True))
        self.d = torch.nn.Parameter(torch.randn(1, requires_grad=True))

    def forward(self, t):
        return self.a * t**3 + self.b * t**2 + self.c * t + self.d

# Initialize model
model = PolyModel()
optimizer = optim.Adam(model.parameters(), lr=learning_rate) 
loss_fn = torch.nn.MSELoss()

loss_history = []
for epoch in range(epochs):
    optimizer.zero_grad()
    N_pred = model(t_tensor)
    loss = loss_fn(N_pred, N_tensor)
    loss.backward()
    optimizer.step()

    loss_history.append(loss.item())

    # Print progress every 50,000 epochs
    if (epoch + 1) % 50000 == 0:
        print(f"Epoch {epoch+1}: Loss = {loss.item():.6f}")

# Q1 Plot loss vs Epoch
plt.figure(figsize=(6, 4))
plt.plot(loss_history)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss Convergence")
plt.show()

# Q2
t_test = np.linspace(t_min, t_max, 500)  
t_test_tensor = torch.tensor(t_test, dtype=torch.float32).view(-1, 1)
N_pred_test = model(t_test_tensor).detach().numpy()  

plt.figure(figsize=(6, 4))
plt.plot(t_values, N_analytical, label="Analytical Solution", linestyle="dashed")
plt.plot(t_test, N_pred_test, label="Fitted Polynomial", color="red")
plt.xlabel("Time (t)")
plt.ylabel("N(t)")
plt.legend()
plt.title("Radioactive Decay: Model vs. Analytical")
plt.show()

# Q3
t_half = np.log(2) / k  
N_half_analytical = N0 * np.exp(-k * t_half)  

t_half_tensor = torch.tensor([[t_half]], dtype=torch.float32)
N_half_pred = model(t_half_tensor).item()  

print(f"Analytical Half-life: t_1/2 = {t_half:.4f} seconds")
print(f"Analytical N(t_1/2) = {N_half_analytical:.4f}")
print(f"Predicted N(t_1/2) = {N_half_pred:.4f}")

# Q4
t_future = torch.tensor([[5000.0]], dtype=torch.float32)
N_future_pred = model(t_future).item()
print(f"\nPrediction at t = 5000.0:")
print(f"Predicted N(5000) = {N_future_pred:.4f}")
```
